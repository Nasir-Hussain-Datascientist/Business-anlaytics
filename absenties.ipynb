```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load your dataset (replace with the actual path of your dataset)
file_name = "HR Help Desk.xlsx"
df = pd.read_excel(file_name)
# Assuming 'df' is your DataFrame, and it's already loaded with the necessary data.

# Filter the DataFrame based on the specified categories in the 'Category Name' column
categories = ['Leave', 'Maternity Leave', 'Paternity Leave', 'Special Leave', 'Carers Leave', 'Bereavement Leave', 'Sickness Absence']
filtered_df = df[df['Category Name'].isin(categories)]
print(filtered_df)
print(filtered_df.describe())
```

```python

# Plot 1: Stacked Bar Chart (X-axis = SUM OF "# of Requests", Y-axis = Category Name)
plot1_data = filtered_df.groupby('Category Name')['# of Requests'].sum().sort_values()
plt.figure(figsize=(10, 6))
plot1_data.plot(kind='barh', stacked=True, color='skyblue')
plt.xlabel('SUM OF # of Requests')
plt.ylabel('Category Name')
plt.title('Stacked Bar Chart: SUM OF # of Requests by Category Name')
plt.show()
print(plot1_data)
print(plot1_data.describe())
```

```python

# Plot 2: Stacked Bar Chart (X-axis = SUM OF "# of Requests", Y-axis = Department Name)
plot2_data = filtered_df.groupby('Department Name')['# of Requests'].sum().sort_values()
plt.figure(figsize=(10, 6))
plot2_data.plot(kind='barh', stacked=True, color='lightgreen')
plt.xlabel('SUM OF # of Requests')
plt.ylabel('Department Name')
plt.title('Stacked Bar Chart: SUM OF # of Requests by Department Name')
plt.show()
print(plot2_data)
print(plot2_data.describe())
```

```python

# Plot 3: Line Chart (Y-axis = SUM OF "# of Requests", X-axis = Creation Date)
filtered_df['Creation Date'] = pd.to_datetime(filtered_df['Creation Date'])  # Ensure dates are in datetime format
plot3_data = filtered_df.groupby('Creation Date')['# of Requests'].sum().sort_index()
plt.figure(figsize=(10, 6))
plt.plot(plot3_data.index, plot3_data.values, marker='o', color='orange', linestyle='-')
plt.xlabel('Creation Date')
plt.ylabel('SUM OF # of Requests')
plt.title('Line Chart: SUM OF # of Requests by Creation Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print(plot3_data)
print(plot3_data.describe())
```
