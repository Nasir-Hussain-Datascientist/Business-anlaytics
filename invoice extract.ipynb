```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For stacked visuals, seaborn or matplotlib can be used.
# Load data from the file
# Load data from the Excel file
file_path = 'Invoice_Extract.xlsx'
df = pd.read_excel(file_path)  # Change 'Sheet1' to the actual sheet name if different

```

```python
# Group data for plotting
df_grouped = df.groupby('Hold Reason').sum().reset_index()

# Plot the stacked column chart
plt.figure(figsize=(10, 6))
plt.bar(df_grouped['Hold Reason'], df_grouped['Amount'], color='skyblue')
plt.xlabel('Hold Reason')
plt.ylabel('Sum of Amount')
plt.title('Stacked Column Chart: Hold Reason vs Sum of Amount')
plt.xticks(rotation=45)
plt.show()

# Print statistics and values
print("Stacked Column Chart Stats:")
print(df_grouped.describe())
print("\nGrouped Data for Plot:")
print(df_grouped)
```

```python
# Group data for plotting
df_grouped = df.groupby('Assigned To').sum().reset_index()

# Plot the stacked bar chart
plt.figure(figsize=(10, 6))
plt.barh(df_grouped['Assigned To'], df_grouped['Amount'], color='lightgreen')
plt.ylabel('Assigned To')
plt.xlabel('Sum of Amount')
plt.title('Stacked Bar Chart: Assigned To vs Sum of Amount')
plt.show()

# Print statistics and values
print("Stacked Bar Chart Stats:")
print(df_grouped.describe())
print("\nGrouped Data for Plot:")
print(df_grouped)
```

```python
# Ensure Due Date is in datetime format
df['Due Date'] = pd.to_datetime(df['Due Date'])

# Group data for plotting (assuming daily granularity)
df_grouped = df.groupby('Due Date').sum().reset_index()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['Due Date'], df_grouped['Amount'], marker='o', linestyle='-')
plt.xlabel('Due Date')
plt.ylabel('Sum of Amount')
plt.title('Line Chart: Due Date vs Sum of Amount')
plt.xticks(rotation=45)
plt.show()

# Print statistics and values
print("Line Chart Stats:")
print(df_grouped.describe())
print("\nGrouped Data for Plot:")
print(df_grouped)
```

```python
# Group data for plotting
df_grouped = df.groupby('Invoice Number').sum().reset_index()

# Plot the stacked column chart
plt.figure(figsize=(10, 6))
plt.bar(df_grouped['Invoice Number'], df_grouped['Amount'], color='coral')
plt.xlabel('Invoice Number')
plt.ylabel('Sum of Amount')
plt.title('Stacked Column Chart: Invoice Number vs Sum of Amount')
plt.xticks(rotation=45)
plt.show()

# Print statistics and values
print("Stacked Column Chart (Invoice Number) Stats:")
print(df_grouped.describe())
print("\nGrouped Data for Plot:")
print(df_grouped)
```
