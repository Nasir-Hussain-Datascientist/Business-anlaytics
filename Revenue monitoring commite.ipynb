```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = "Revenue Monitoring Committee Report April.xlsx"
sheet_name = "Sheet1"  # Update with your sheet name
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Trim any extra spaces in column names
data.columns = data.columns.str.strip()

# Top 5 Cost Centers by Revised Budget 
top_5_revised_budget = data.groupby('Cost Centres')['Revised\nBudget '].sum().nlargest(5)

# Plot 1
plt.figure(figsize=(10, 6))
top_5_revised_budget.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Cost Centres')
plt.ylabel('Sum of Revised Budget ')
plt.title('Top 5 Cost Centres by Sum of Revised Budget ')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Print values
print("Top 5 Cost Centres by Sum of Revised Budget :")
print(top_5_revised_budget)
```

```python
# Bottom 5 Cost Centres by Revised Budget 
bottom_5_revised_budget = data.groupby('Cost Centres')['Revised\nBudget '].sum().nsmallest(5)

# Plot 2
plt.figure(figsize=(10, 6))
bottom_5_revised_budget.plot(kind='bar', color='salmon', edgecolor='black')
plt.xlabel('Cost Centres')
plt.ylabel('Sum of Revised Budget ')
plt.title('Bottom 5 Cost Centres by Sum of Revised Budget ')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Print values
print("Bottom 5 Cost Centres by Sum of Revised Budget :")
print(bottom_5_revised_budget)
```

```python
# Top 5 Cost Centres by Full Year Variance
top_5_full_year_variance = data.groupby('Cost Centres')['Full Year Variance'].sum().nlargest(5)

# Plot 3
plt.figure(figsize=(10, 6))
top_5_full_year_variance.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Cost Centres')
plt.ylabel('Sum of Full Year Variance')
plt.title('Top 5 Cost Centres by Sum of Full Year Variance')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Print values
print("Top 5 Cost Centres by Sum of Full Year Variance:")
print(top_5_full_year_variance)
```

```python
# Bottom 5 Cost Centres by Full Year Variance
bottom_5_full_year_variance = data.groupby('Cost Centres')['Full Year Variance'].sum().nsmallest(5)

# Plot 4
plt.figure(figsize=(10, 6))
bottom_5_full_year_variance.plot(kind='bar', color='salmon', edgecolor='black')
plt.xlabel('Cost Centres')
plt.ylabel('Sum of Full Year Variance')
plt.title('Bottom 5 Cost Centres by Sum of Full Year Variance')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Print values
print("Bottom 5 Cost Centres by Sum of Full Year Variance:")
print(bottom_5_full_year_variance)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = "Revenue Monitoring Committee Report April.xlsx"
sheet_name = "Sheet1"  # Update with your sheet name
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Trim any extra spaces in column names
data.columns = data.columns.str.strip()

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(data['% Variance'], data['Full Year Variance'], color='dodgerblue', edgecolor='black')
plt.xlabel('% Variance')
plt.ylabel('Full Year Variance')
plt.title('Scatter Plot of % Variance vs Full Year Variance')
plt.grid(True)
plt.tight_layout()
plt.show()

# Print descriptive statistics
print("Descriptive Statistics for % Variance:")
print(data['% Variance'].describe())

print("\nDescriptive Statistics for Full Year Variance:")
print(data['Full Year Variance'].describe())
```
