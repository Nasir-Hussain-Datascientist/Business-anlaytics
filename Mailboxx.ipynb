```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
file_path = "Mailboxx.xlsx"
df = pd.read_excel(file_path, sheet_name=None)

# Extract the data from the relevant sheet
data = df[list(df.keys())[0]]

# Plot 1: Line Chart - SUMOF Accounts Payable Mailbox
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Accounts Payable Mailbox'], marker='o', color='blue', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Sum of Accounts Payable Mailbox')
plt.title('Line Chart: Date vs Sum of Accounts Payable Mailbox')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Print all data for Plot 1
print("Line Chart (Accounts Payable Mailbox) Values:")
print(data[['Date', 'Accounts Payable Mailbox']])

# Print statistics for Plot 1
print("\nLine Chart (Accounts Payable Mailbox) Statistics:")
print(data['Accounts Payable Mailbox'].describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
file_path = "Mailboxx.xlsx"
df = pd.read_excel(file_path, sheet_name=None)

# Extract the data from the relevant sheet
data = df[list(df.keys())[0]]

# Summarize data for Pie Chart
summary = {
    'Accounts Payable Mailbox': data['Accounts Payable Mailbox'].sum(),
    'Creditors Creations Mailbox': data['Creditors Creations Mailbox'].sum(),
    'Creditors Mailbox': data['Creditors Mailbox'].sum(),
    'Creditors Manual Mailbox': data['Creditors Manual Mailbox'].sum()
}

# Plot 3: Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(summary.values(), labels=summary.keys(), autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'lightskyblue'])
plt.title('Pie Chart: Sum of Mailbox Categories')
plt.show()

# Print all data for Plot 3
print("Pie Chart Values:")
print(summary)

# Print statistics for Plot 3
print("\nPie Chart (Mailbox Categories) Statistics:")
for category, value in summary.items():
    print(f"{category}: {value}")
```
