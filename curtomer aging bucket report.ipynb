```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
file_path = "SC AR Customer Aging by Bucket List Report_Output - Anonymous.xlsx"
data = pd.read_excel(file_path, sheet_name='Sheet1')


# Prepare data for plotting
data_grouped = data.groupby('Aging Bucket')['Original Balance'].sum().reset_index()

# Print Descriptive Statistics
desc_stats = data_grouped['Original Balance'].describe()
print("Descriptive Statistics for Original Balance by Aging Bucket:")
print(desc_stats)
print("\n")
print(data_grouped)
print("\n")


# Plot Stacked Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.bar(data_grouped['Aging Bucket'], data_grouped['Original Balance'])
plt.xlabel('Aging Bucket')
plt.ylabel('SUMOF Original Balance')
plt.title('Aging Bucket vs. SUMOF Original Balance')
plt.xticks(rotation=45)

# Annotate values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.2f}', va='bottom', ha='center')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data

# Prepare data for plotting
top_customers = data.groupby('Customer Name')['Amount Remaining'].sum().nlargest(10).reset_index()

# Print Descriptive Statistics
desc_stats = top_customers['Amount Remaining'].describe()
print("Descriptive Statistics for Amount Remaining by Top 10 Customers:")
print(desc_stats)
print("\n")
print(top_customers)
# Plot Clustered Column Chart
plt.figure(figsize=(12, 8))
bars = plt.bar(top_customers['Customer Name'], top_customers['Amount Remaining'])
plt.xlabel('Customer Name')
plt.ylabel('SUMOF Amount Remaining')
plt.title('Top 10 Customers by Amount Remaining')
plt.xticks(rotation=45)

# Annotate values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.2f}', va='bottom', ha='center')

plt.tight_layout()
plt.show()
```

```python
import squarify
# Prepare data for plotting
top_profiles = data.groupby('Profile Class')['Original Balance'].sum().nlargest(5).reset_index()

# Print Descriptive Statistics
desc_stats = top_profiles['Original Balance'].describe()
print("Descriptive Statistics for Original Balance by Top 5 Profiles:")
print(desc_stats)
print("\n")
print(top_profiles)
# Plot Treemap
plt.figure(figsize=(10, 8))
squarify.plot(sizes=top_profiles['Original Balance'], label=top_profiles['Profile Cass'], alpha=.8, color=plt.cm.viridis(top_profiles['Original Balance'] / max(top_profiles['Original Balance'])))
plt.title('Top 5 Profiles by SUMOF Original Balance')
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Prepare data for plotting
data_grouped = data.groupby('Aging Bucket')['Original Balance'].sum().reset_index()

# Print Descriptive Statistics
desc_stats = data_grouped['Original Balance'].describe()
print("Descriptive Statistics for Original Balance by Aging Bucket:")
print(desc_stats)
print("\n")
print(data_grouped)

# Plot Pie Chart
plt.figure(figsize=(8, 8))
sizes = data_grouped['Original Balance']
labels = data_grouped['Aging Bucket']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Original Balance by Aging Bucket')
plt.axis('equal')
plt.show()
```

```python

# Prepare data for plotting
data_grouped = data.groupby('Due Date').agg({
    'Collected': 'sum',
    'Original Balance': 'sum'
}).reset_index()

# Print Descriptive Statistics
desc_stats = data_grouped[['Collected', 'Original Balance']].describe()
print("Descriptive Statistics for Collected and Original Balance by Due Date:")
print(desc_stats)
print("\n")
print(data_grouped)
# Plot Line Chart
plt.figure(figsize=(12, 8))
plt.plot(data_grouped['Due Date'], data_grouped['Collected'], label='SUMOF Collected', marker='o')
plt.plot(data_grouped['Due Date'], data_grouped['Original Balance'], label='SUMOF Original Balance', marker='o')
plt.xlabel('Due Date')
plt.ylabel('Amount')
plt.title('Due Date vs. SUMOF Collected and SUMOF Original Balance')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# Annotate values
for i in range(len(data_grouped)):
    plt.text(data_grouped['Due Date'][i], data_grouped['Collected'][i], f'{data_grouped["Collected"][i]:.2f}', fontsize=9, ha='right')
    plt.text(data_grouped['Due Date'][i], data_grouped['Original Balance'][i], f'{data_grouped["Original Balance"][i]:.2f}', fontsize=9, ha='right')

plt.tight_layout()
plt.show()
```
