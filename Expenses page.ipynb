```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
file_path = "All Expenses.xlsx"
data = pd.read_excel(file_path)
```

```python
# Group data by Expense Type and sum the reimbursable amount
expense_type_data = data.groupby('Expense Type')['Reimbursable Amount'].sum()

# Pie chart
plt.figure(figsize=(8,8))
plt.pie(expense_type_data, labels=expense_type_data.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Expense Type vs Sum of Reimbursable Amount')
plt.show()

# Print the values used for the pie chart
print("Pie Chart Values (Expense Type vs Sum of Reimbursable Amount):\n", expense_type_data)

# Summary
summary_pie = expense_type_data.describe()
print("Summary of Pie Chart (Expense Type vs Sum of Reimbursable Amount):\n", summary_pie)
```

```python
# Scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(x=data['Trip Distance'], y=data['Reimbursable Amount'])
plt.title('Trip Distance vs Sum of Reimbursable Amount')
plt.xlabel('Trip Distance')
plt.ylabel('Reimbursable Amount')
plt.show()

# Print the values used for the scatter plot
print("Scatter Plot Values (Trip Distance vs Reimbursable Amount):")
print(data[['Trip Distance', 'Reimbursable Amount']])

# Summary
summary_scatter = data[['Trip Distance', 'Reimbursable Amount']].describe()
print("Summary of Scatter Plot (Trip Distance vs Reimbursable Amount):\n", summary_scatter)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

# Load your data
# data = pd.read_csv('your_file.csv')  # Replace with your actual file path

# Convert 'Employee Name' to string format
data['Employee Name'] = data['Employee Name'].astype(str)

# Group data by Employee Name and sum the reimbursable amount
employee_data = data.groupby('Employee Name')['Reimbursable Amount'].sum()

# Filter out zero or negative values
employee_data = employee_data[employee_data > 0]

# Check if data is empty
if employee_data.empty:
    print("No data to plot. Filtered dataset is empty.")
else:
    # Treemap
    plt.figure(figsize=(12, 8))
    try:
        squarify.plot(sizes=employee_data.values, label=employee_data.index, alpha=.8)
        plt.title('Employee Name vs Sum of Reimbursable Amount')
        plt.axis('off')
        plt.show()
    except ZeroDivisionError as e:
        print("Error plotting treemap:", e)

    # Print the values used for the treemap
    print("Treemap Values (Employee Name vs Sum of Reimbursable Amount):\n", employee_data)

    # Summary
    summary_treemap = employee_data.describe()
    print("Summary of Treemap (Employee Name vs Sum of Reimbursable Amount):\n", summary_treemap)
```

```python
# Convert Final Approval Date to month format
data['Final Approval Month'] = pd.to_datetime(data['Final Approval Date']).dt.to_period('M')

# Group by month and sum the reimbursable amount
monthly_data = data.groupby('Final Approval Month')['Reimbursable Amount'].sum()

# Line chart
plt.figure(figsize=(10,6))
monthly_data.plot(kind='line', marker='o', color='b')
plt.title('Sum of Reimbursable Amount by Final Approval Month')
plt.xlabel('Month')
plt.ylabel('Reimbursable Amount')
plt.xticks(rotation=45)
plt.show()

# Print the values used for the line chart
print("Line Chart Values (Final Approval Month vs Sum of Reimbursable Amount):\n", monthly_data)

# Summary
summary_line = monthly_data.describe()
print("Summary of Line Chart (Final Approval Date vs Sum of Reimbursable Amount):\n", summary_line)
```

```python
# Convert Cost Center Code to text if needed
data['Cost Center Code'] = data['Cost Center Code'].astype(str)

# Group data by Cost Center Code and sum the reimbursable amount
cost_center_data = data.groupby('Cost Center Code')['Reimbursable Amount'].sum()

# Stacked bar chart
plt.figure(figsize=(12,6))
cost_center_data.plot(kind='bar', stacked=True)
plt.title('Cost Center Code vs Sum of Reimbursable Amount')
plt.xlabel('Cost Center Code')
plt.ylabel('Reimbursable Amount')
plt.show()

# Print the values used for the bar chart
print("Stacked Bar Chart Values (Cost Center Code vs Sum of Reimbursable Amount):\n", cost_center_data)

# Summary
summary_bar = cost_center_data.describe()
print("Summary of Stacked Bar Chart (Cost Center Code vs Sum of Reimbursable Amount):\n", summary_bar)
```
