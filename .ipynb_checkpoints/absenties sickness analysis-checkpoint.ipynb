```python
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load your dataset (replace with the actual path of your dataset)
file_name = "HR Help Desk.xlsx"
df = pd.read_excel(file_name)

# Filtering data by Category Name = "Sickness Absence"
df_filtered = df[df['Category Name'] == 'Sickness Absence']

# Calculate Working Days Lost
def calculate_working_days_lost(row):
    if pd.notnull(row['Creation Date']) and pd.notnull(row['Resolution Date']):
        start_date = row['Creation Date']
        end_date = row['Resolution Date']
        # Create a date range from start to end date
        date_range = pd.date_range(start=start_date, end=end_date, freq='B')  # 'B' excludes weekends
        return len(date_range)  # Return the number of working days
    else:
        return None

# Apply the calculation to create a new column
df_filtered['Working_Days_Lost'] = df_filtered.apply(calculate_working_days_lost, axis=1)

# Summary statistics for the 'Working_Days_Lost' column
print(df_filtered['Working_Days_Lost'].describe())
print(df_filtered)

# Plot 1: Stacked Bar Chart (y axis = Department Name, x axis = SUM OF Working_Days_Lost)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_filtered, x='Working_Days_Lost', y='Department Name', estimator=sum, ci=None)
plt.title('Stacked Bar Chart: Department Name vs SUM OF Working Days Lost')
plt.xlabel('Sum of Working Days Lost')
plt.ylabel('Department Name')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Line Chart (x axis = Creation Date, y axis = SUM OF Working_Days_Lost)
df_filtered['Creation Date'] = pd.to_datetime(df_filtered['Creation Date'])
working_days_by_date = df_filtered.groupby('Creation Date')['Working_Days_Lost'].sum().reset_index()
print(working_days_by_date )
print(working_days_by_date.describe())
plt.figure(figsize=(10, 6))
sns.lineplot(data=working_days_by_date, x='Creation Date', y='Working_Days_Lost')
plt.title('Line Chart: Creation Date vs SUM OF Working Days Lost')
plt.xlabel('Creation Date')
plt.ylabel('Sum of Working Days Lost')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate the SUM of # of Requests by Department Name
grouped_data = df_filtered.groupby('Department Name')['# of Requests'].sum().reset_index()

# Print the data for the graph
print("\nData for Stacked Bar Chart (Department Name vs SUM OF # of Requests):")
print(grouped_data)

# Summary statistics for the '# of Requests' column
print("\nSummary statistics for '# of Requests':")
print(df_filtered['# of Requests'].describe())

# Plot 3: Stacked Bar Chart (y axis = Department Name, x axis = SUM OF "# of Requests")
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='# of Requests', y='Department Name', estimator=sum, ci=None)
plt.title('Stacked Bar Chart: Department Name vs SUM OF "# of Requests"')
plt.xlabel('Sum of # of Requests')
plt.ylabel('Department Name')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import squarify

# Prepare data for the tree map
tree_map_data = df_filtered.groupby('Employee Assignment Number')['# of Requests'].sum().reset_index()
print(tree_map_data)
print(tree_map_data.describe())
# Plot the tree map
plt.figure(figsize=(10, 6))
squarify.plot(sizes=tree_map_data['# of Requests'], label=tree_map_data['Employee Assignment Number'], alpha=0.9)
plt.axis('off')
plt.title('Tree Map: Employee Assignment Number vs SUM OF "# of Requests"')
plt.tight_layout()
plt.show()
```
